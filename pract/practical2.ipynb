```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/temperatures.csv')
```

```python
df
```

```python
#input data 
x = df['YEAR']

#output data
y = df['ANNUAL']
```

```python
plt.title = ('Temperature plot of INDIA')
plt.xlabel('YEAR')
plt.ylabel('ANNUAL Average Temperature')
plt.scatter(x , y)
```

```python
x.shape
```

```python
x = x.values
```

```python
x = x.reshape(117, 1)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[2022]])
```

```python
predicted = regressor.predict(x)
```

```python
import numpy as np
```

```python
#mean absolute error
np.mean(abs(y - predicted))
```

```python
#mean square error
np.mean(abs(y - predicted) **2)
```

```python
#r square matrix
regressor.score(x, y)
```

```python
plt.title = ('Temperature plot of INDIA')
plt.xlabel('YEAR')
plt.ylabel('ANNUAL Average Temperature')
plt.scatter(x , y , label = 'actual' , color = 'r')
plt.plot(x , predicted , label = 'predicted' , color = 'g')
plt.legend()
```
